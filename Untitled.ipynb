```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_excel('https://github.com/anastasiarazb/skillbox-data-analyst-intensive/blob/master/%D0%90%D0%BD%D0%B0%D0%BB%D0%B8%D0%B7%20%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D1%85%20%D0%94%D0%B5%D0%BD%D1%8C%202/IPG2211A2N.xls?raw=true', skiprows=10)
```

```
data_daily['IPG2211A2N'].quantile(0.95)
```

```
# размножим значение, чтобы получить столбец и отрисовать его на графике
N = len(data_daily) # количество записей
# Список длины N со скопированным значением квантиля - размножили список из 1 элемента N раз
quant_95 = [data_daily['IPG2211A2N'].quantile(0.95)] * N  # 95%
quant_05 = [data_daily['IPG2211A2N'].quantile(0.05)] * N  # 5%
```

```
plt.plot(data_daily.index, data_daily['IPG2211A2N']);
plt.plot(data_daily.index, quant_95);
plt.plot(data_daily.index, quant_05);
```

```
pd.plotting.autocorrelation_plot(data_daily['IPG2211A2N']);
```

```
plt.figure(figsize=(20, 5))  # размер графика
pd.plotting.autocorrelation_plot(data_daily['IPG2211A2N'])
plt.locator_params(axis='x', nbins=50);
```

```
from sklearn.linear_model import LinearRegression
```

```
regressor = LinearRegression()
```

```
x_range = list(range(N))
X = pd.DataFrame(x_range)
```

```
regressor.fit(X, data_daily['IPG2211A2N'])
```

```
result = regressor.predict(X)
```

```
plt.plot(X.index, data_daily['IPG2211A2N'])
plt.plot(X.index, result);
```
